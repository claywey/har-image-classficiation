```python
from typing import Tuple

import pandas as pd 
import numpy as np

from tqdm import tqdm
from pathlib import Path

import asyncio
import os

from PIL import Image
from IPython.display import display

from tensorflow.keras.utils import image_dataset_from_directory
```

```python
BASE_PATH_TO_DATA = "../input/human-action-recognition-har-dataset/Human Action Recognition/"
```

```python
def load_labels() -> Tuple[pd.DataFrame]:
    """Utility to load the labeled data"""
    train_labels = pd.read_csv(os.path.join(BASE_PATH_TO_DATA, "Training_set.csv"))
    test_labels = pd.read_csv(os.path.join(BASE_PATH_TO_DATA, "Testing_set.csv"))
    
    return (train_labels, test_labels)

async def load_images(labels: pd.DataFrame, dataset: str) -> np.array:
    images = np.array([], dtype = np.float32)
    size = len(labels)
    
    for indx in tqdm(range(size)):
        path = os.path.join(BASE_PATH_TO_DATA, dataset, labels.loc[indx].filename)
        image = await Image.open(path)
        images = np.append(images, np.asarray(image.resize((160,160))))
    return images

def create_keras_dataset():
    train_labels, test_labels = load_labels()
    train_imgs = load_images(train_labels, "train")
    test_imgs = load_images(test_labels, "test")
    
```
